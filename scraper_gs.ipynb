```python
import time
import json

from scholarly import scholarly
from scholarly import ProxyGenerator

from icecream import ic
```

```python
# initialise proxy generator

pg = ProxyGenerator()

# add proxy as a method - this should return True is proxy was set up successfully

# TODO: replace 'YOUR_API obtained through https://www.scraperapi.com/ - this is a paid service

success = pg.ScraperAPI('YOUR_API')
print(success)

# use proxy
scholarly.use_proxy(pg,pg)
```

```python
language = 'portuguese'
i = '2022'
j = 'por22'
_ = '2'

# TODO: replace 'GoogleScholar_URL' with the URL of the search results

queries = {
    "por_2021": "GoogleScholar_URL",
    "por_2022": "GoogleScholar_URL",
}


article_count = 0

# TODO: replace 'YOUR_PATH'
 
with open(f'./YOUR_PATH/{language}/query_{_}/articles_{language}_year_{i}.json', 'w', encoding='utf-8') as file:
    exports = scholarly.search_pubs_custom_url(f'{queries[j]}')

    rows = []

    for hit in exports:
        title = hit["bib"]["title"]
        venue = hit["bib"]["venue"] # journal name
        year = hit["bib"]["pub_year"]

        journal = venue if venue != "NA" else None
        
        url = None
        if "eprint_url" in hit:
            url = hit["eprint_url"]
        elif "pub_url" in hit:
            url = hit["pub_url"]
        
        authors = ", ".join(hit["bib"]["author"])
        pub_year = str(year) if year.isdecimal() else None

        rows.append({
            "T1": title,
            "AU": authors,
            "LA": language,
            "PY": pub_year,
            "JO": journal,
            "DO": url
        })

        article_count += 1
        ic(article_count)
        
        # if article_count % 20 == 0:
        #     time.sleep(5)
   
    json.dump(rows, file) # save dictionary as json file - dump(object to write, file to write object into)
    ic(i, "done", len(rows))
    
```
